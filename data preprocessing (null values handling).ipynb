```python
x=ds[['new_tests']].values
x
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(x[:])
x_imputed = imputer.transform(x[:])

# Create a new dataframe with the imputed values
ds_imputed = pd.DataFrame(x_imputed)
# Concatenate the original dataframe and the imputed dataframe
ds_final = pd.concat([ds, ds_imputed], axis=1)
```

```python
ds=ds_final.rename(columns={0:'new_test'})
```
